## Mimic the tossing of a biased/unfair coin with P(head) = 1/4 and P(tail) = 3/4

```python
import numpy as np 
head = 0
tail = 0
for i in range(1000):
    toss = np.random.randint(1,101)
    if toss<=25:
        head+=1
    else:
        tail+=1
print(f"Ratio of heads:{head/1000*100:.2f}%")
print(f"Ratio of tails:{tail/1000*100:.2f}%")
```

## Multivariate normal distribution

```python
import numpy as np 
mean = [11,13]
cov = [[40,30],[30,50]]
multi = np.random.multivariate_normal(mean,cov,5000)
print(multi)
import matplotlib.pyplot as plt 
plt.scatter(multi[:,0], multi[:,1])
plt.show()
```

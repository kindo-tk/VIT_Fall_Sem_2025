## 1- Generate random head and tail, repeat the experiment 10000 times and find the ratio of head and tail

```python
## Head - 1
## Tail - 0

import numpy as np
head = 0
tail = 0 
for i in range(10000):
    toss = np.random.randint(0,2)
    if toss == 1:
        head+=1
    else:
        tail+=1
print(f"Proportion of head:{head/10000*100}%")
print(f"Proportion of head:{tail/10000*100}%")
```

## 2- Experiment is to toss a fair coin 10 times, x is the event of occurrence of heads then run this experiment 10000 times.


```python
import numpy as np 
head_count = []
for j in range(10000):
    head = 0
    for i in range(10):
        toss = np.random.randint(0,2)
        if toss == 1:
            head += 1
    head_count.append(head)
print("Head count:", head_count)

d = {}
for i in head_count:
    if i not in d:
        d[i] = 1
    else:
        d[i] += 1
for i in range(11):
    if i not in d:
        d[i] = 0
d = dict(sorted(d.items()))
print("\nNumber of success in 10000 trials:\n",d)
for key in d.keys():
    print(f"\nProportion(x = {key})={d[key]/10000}")
```
